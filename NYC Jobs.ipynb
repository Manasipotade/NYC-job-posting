```python
cols = df.columns
num_cols = df._get_numeric_data().columns
date_cols = df.select_dtypes(include=['datetime64']).columns
cat_cols = list(set(cols) - set(num_cols) - set(date_cols))
```

### Remove StopWords

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
def remove_stopwords(input_text):
    new_text = " ".join([i for i in input_text if i not in stop_words])
    return new_text
df['Work Location'] = df['Work Location'].apply(lambda x : remove_stopwords(x.split()))
df['Residency Requirement'] = df['Residency Requirement'].apply(lambda x : remove_stopwords(x.split()))
df['Job Description'] = df['Job Description'].apply(lambda x : remove_stopwords(x.split()))
df['To Apply'] = df['To Apply'].apply(lambda x : remove_stopwords(x.split()))
df['Additional Information'] = df['Additional Information'].apply(lambda x : remove_stopwords(x.split()))
# Residency Requirement
```
